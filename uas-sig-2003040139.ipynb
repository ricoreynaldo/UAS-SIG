```python
import geopandas as gpd
#mengimport modul geopandas sebagai gpd
```

```python
gdf = gpd.read_file('https://raw.githubusercontent.com/ricoreynaldo/UAS-SIG/main/RBI25K_ADMINISTRASI_AR.geojson')
#membaca file geojson dari github dan disimpan dalam variabel gdf
gdf.head()
#menampilkan 5 baris pertama dari variabel gdf
```

```python
gdf = gdf[['NAMOBJ', 'FCODE', 'REMARK', 'METADATA', 'geometry']].fillna(0)
#memilih kolom 'NAMOBJ', 'FCODE', 'REMARK', 'METADATA', 'geometry' dan menggantikan nilainya menjadi 0
```

```python
!pip install folium matplotlib mapclassify
#menginstal modul folium, matplotlib dan maoclassify
```

```python
import folium
#mengimport modul folium
m = gdf.explore(column="METADATA", name="Metadata")
#membuat map dengan folium dan disimpan ke variabel m
folium.LayerControl().add_to(m)
#menambahkan layer control ke objek map dengan variabel m
m
#menampilkan map
```
